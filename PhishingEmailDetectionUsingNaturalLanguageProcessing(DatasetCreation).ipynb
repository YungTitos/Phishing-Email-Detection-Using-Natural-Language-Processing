## Modules

```python
import pandas as pd
```

## Corpus Creation

```python
processed_dfs = []

# Nazario Dataset
df_naz = pd.read_csv("Nazario.csv")
df_naz = df_naz[['body', 'label']].rename(columns={'body': 'email'})
df_naz['source'] = 'Nazario'
processed_dfs.append(df_naz)

# Nigerian Fraud Dataset
df_nig = pd.read_csv("Nigerian_Fraud.csv")
df_nig = df_nig[['body', 'label']].rename(columns={'body': 'email'})
df_nig['source'] = 'Nigerian_Fraud'
processed_dfs.append(df_nig)

# KD_10000 Dataset
df_kd = pd.read_csv("phishing_legit_dataset_KD_10000.csv")
df_kd = df_kd[['text', 'label']].rename(columns={'text': 'email'})
df_kd['source'] = 'KD_10000_Synthetic'
processed_dfs.append(df_kd)

# Twente Dataset
df_val = pd.read_csv("Phishing_validation_emails.csv")
label_map = {"Safe Email": 0, "Phishing Email": 1}
df_val['label'] = df_val['Email Type'].map(label_map)
df_val = df_val[['Email Text', 'label']].rename(columns={'Email Text': 'email'})
df_val['source'] = 'Twente_2024'
processed_dfs.append(df_val)

# Enron Dataset
df_enron = pd.read_csv("Enron.csv")
df_enron['email'] = df_enron['subject'].fillna('') + "\n" + df_enron['body'].fillna('')
df_enron = df_enron[['email', 'label']]
df_enron['source'] = 'Enron'
processed_dfs.append(df_enron)

unified_df = pd.concat(processed_dfs, ignore_index=True)

unified_df = unified_df.dropna(subset=['email'])
unified_df.drop_duplicates(inplace=True)
unified_df['label'] = unified_df['label'].astype(int)

# Save the final result
unified_df.to_csv("phishing_email_corpus.csv", index=False)
print("☻")
```

## Mini Report

The corpus utilizes the *Enron* dataset as the primary baseline for legitimate correspondence, providing a realistic profile of professional corporate discourse. To model deceptive tactics, the *Nazario* and *Nigerian Fraud* datasets are included for their focus on classic social engineering and advance-fee fraud.

To bridge temporal gaps, the *KD_10000* Synthetic dataset introduces contemporary lures, such as cloud-service authentication and cryptocurrency-related threats. Finally, the *University of Twente* dataset is integrated as real-world and ai-assisted phishing samples to help generalize the dataset.

The preprocessing pipeline standardized disparate schemas into a unified "email" and "label" format. For the Enron source, subject and body fields were concatenated to preserve context. Labels were mapped to a binary system (Safe: 0, Phishing: 1), followed by the removal of null entries and strict deduplication to ensure data integrity. The resulting corpus provides a multi-generational training environment, that balances organic corporate "noise" with evolving fraudulent architectural features.
